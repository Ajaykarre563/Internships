# **what is mean by power BI ?**

**Power BI is a business intelligence tool from Microsoft that helps people turn raw data into useful insights.**

 **It allows you to:**

Connect to data from different sources (like Excel, databases, or online services).

Visualize data by creating interactive charts, graphs, and dashboards.

Share insights with others through reports and dashboards.

**In simple terms, Power BI helps you see and understand your data clearly so you can make better decisions. It's like a smart, interactive report generator.**

# **Main Types of Power BI (Simplified with an Example):**

**Power BI Desktop**


This is the tool you use to create reports and visuals.
Example: You design a sales report with charts and graphs on your computer.

**Power BI Service**

A cloud platform to share and collaborate on reports.
Example: After creating the sales report in Power BI Desktop, you upload it to the service so your team can view it online.

**Power BI Mobile**

An app to access reports on the go.
Example: While traveling, you open the sales report on your phone to check the latest numbers.

# **TASK:2 CHURN**

**Churn** refers to the loss of customers or subscribers from a business or service over a specific period. It is a key metric for measuring customer retention and understanding business performance.

**In Your Dataset:**

Churn Column: This column indicates whether a customer has left the company.
Yes: The customer has churned (i.e., discontinued their service).
No: The customer is still active and has not churned.

**Example:**

A customer who cancels their internet service or switches to a competitor is considered churned.

**Why Churn is Important:**

**Retention Analysis:**

Understanding churn helps identify why customers leave and what can be done to retain them.

Retaining existing customers is often more cost-effective than acquiring new
ones.

**Business Impact:**

High churn rates can lead to declining revenue and market share.
Churn analysis enables businesses to create targeted strategies for improving customer satisfaction and loyalty.

# **Task 3:Diversity and Inclusion**

**Diversity and Inclusion (D&I) refers to efforts and strategies aimed at creating a workplace where individuals from diverse backgrounds feel valued, respected, and supported.**

**Diversity** focuses on the representation of different groups within an organization, including differences in gender, age, race, ethnicity, religion, disability, sexual orientation, and other characteristics. It ensures that people from all walks of life are included.

**Inclusion** ensures that everyone, regardless of their background, feels welcome, heard, and empowered to contribute fully to the organization. It's about creating an environment where diverse individuals feel they belong.

**Key Goals of Diversity and Inclusion:**

Representation: Achieving diversity across all levels and departments of the organization.

Equity: Providing equal opportunities for all employees to succeed and grow.

Belonging: Creating a culture where every employee feels valued and included.

Innovation: Leveraging diverse perspectives to drive creativity and better decision-making.

Retention: Building an environment that encourages employees to stay and thrive.
