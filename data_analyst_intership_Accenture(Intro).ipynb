# **what is mean by data analyst?**

**A data analyst is someone who examines data to find patterns, trends, or useful insights that can help make decisions. They collect, clean, and analyze data, then present the results in a way that others can understand, like through charts or reports.**

**Example:**

Imagine a company that sells clothes online. A data analyst might:

Look at sales data to find which items are most popular.
Identify what time of year certain clothes sell the most (like jackets in winter).

Suggest stocking more of these items during those times to increase sales.

# **what is Dataset?**

**A dataset is a collection of data organized in a structured way, like a table**.

**It contains rows and columns:**

Rows: Each row represents a single record or item (e.g., one customer, one product, one transaction).

Columns: Each column represents a specific type of information or attribute (e.g., age, price, date).

# **what is mean by Data Model ?**

**A data model shows how different datasets (tables of data) are connected and related. It also shows the links that can be used to combine or merge these tables to make sense of the data as a whole.**

**Example:**

**In a shopping website:**

There’s a Customer dataset (table with customer details).

There’s an Order dataset (table with purchase details).

There’s a Product dataset (table with product details).
A data model would show:

The Customer dataset is linked to the Order dataset by the Customer ID (1 customer can place many orders).

The Order dataset is linked to the Product dataset by the Order ID (1 order can include many products).

# **what is mean by VLOOKUP ?**

**VLOOKUP (Vertical Lookup) is a function in Excel and other spreadsheet tools used to find and retrieve data from a specific column in a table, based on a matching value in another column.**

**VLOOKUP is like looking for a friend's name in a contact list to find their phone number**.

**For example:**

If you have a table with Employee IDs and their salaries, VLOOKUP helps you find the salary by entering the Employee ID.

# **what is mean by SUMIF ?**

**SUMIF in Excel adds numbers based on a condition.**

For example:
If you have a table with sales data and want to add only the sales made by "Alice," SUMIF will calculate the total sales for Alice.

It’s like saying, “Add these numbers, but only if they meet this condition.”

# **Note:**

**SUMIFS in Excel is like SUMIF, but it allows you to add numbers based on multiple conditions.**

**Simple Explanation:**

If SUMIF is “Add the numbers that meet one condition,”
then SUMIFS is “Add the numbers that meet multiple conditions.”

# **what is mean by COUNTIF?**

**COUNTIF is used to count how many times something appears in a list.**

**For example:**

If you have a list of names and want to know how many times "Alice" appears, COUNTIF will count it for you.
